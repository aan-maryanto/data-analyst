```python
import matplotlib.pyplot as plt

X = range(150)
Y = [value ** 2 for value in X]
# Y = range(1000)
plt.plot(X, Y)
plt.show()

```

```python
import math
import matplotlib.pyplot as plt

T = range(100)
X = [(2 * math.pi * t) / len(T) for t in T]
Y = [math.sin(value) for value in X]
plt.plot(X,Y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
X = np.linspace(0, 2 * np.pi, 100)
Y = np.sin(X)

plt.plot(X,Y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(-3, 2, 200)
Y = X ** 2 - 2 * X + 1

plt.plot(X,Y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(0, 2 * np.pi, 100)
Ya = np.sin(X)
Yb = np.cos(X)

plt.plot(X, Ya)
plt.plot(X, Yb)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_slope(X,Y):
    Xs = X[1:] - X[:-1]
    Ys = Y[1:] - Y[:-1]
    plt.plot(X[1:], Ys/Xs)
    
X = np.linspace(-3, 3, 100)
Y = np.exp(-X ** 2)
plt.plot(X,Y)
plot_slope(X,Y)
plt.show()
```

```python
import matplotlib.pyplot as plt

X, Y = [], []
for line in open('my_data.txt', 'r'):
    values = [float(s) for s in line.split()]
    X.append(values[0])
    Y.append(values[1])

plt.plot(X,Y)
plt.show()
```
