```python
import pandas as pd
from pandas_datareader import data
import matplotlib.pyplot as plt
```

```python
bb = data.DataReader("BB", data_source="yahoo", start="2006-01-01", end="2020-01-01")
bb.head(3)
```

```python
bb.plot(y=["High", "Low"])
```

```python
 bb[["Open", "Close"]].plot()
```

```python
plt.style.available
```

```python
plt.style.use('ggplot')
bb.plot(y="Close")
```

```python
def rank_performance(stock_price) :
    if stock_price <= 10:
        return "Poor"
    elif stock_price <= 50:
        return "Satisfactory"
    else:
        return "Stellar"
```

```python
plt.style.use("ggplot")
bb["Close"].apply(rank_performance).value_counts().plot(kind="bar")
```

```python
bb["Close"].mean()
```

```python
def rank_by_mean(stock_price):
    if stock_price >= 31.14440156003815:
        return "Above Average"
    else:
        return "Below Average"
```

```python
plt.style.use("bmh")
bb["Close"].apply(rank_by_mean).value_counts().plot(kind="pie", legend=True)
```
